## Question 2

```python
import numpy as np

def Sine(a,f,p,ts,lbl,clr):
    ps = p * np.pi/180
    t = np.linspace(0,ts,1000)
    v1 = a*np.sin((2*np.pi*f*t)+ps)
    plt.figure(2)
    plt.plot(t,v1, color = clr, label = lbl)
    plt.legend()
    plt.grid()
    
def SetPlot(xlbl,ylbl,tit):
    plt.xlabel(xlbl) # Labels x-axis
    plt.ylabel(ylbl) # label y-axis
    plt.title(tit)
    
plt.figure(1)
Sine(5,5,0,2,"v1",'red')
SetPlot('time','amplitude','Phase A')
```

## Question 3

```python
def PrintSine(A,f,p,ts,lbl):
    def sin_plot(t,v,lbl, tit, clr): 
        plt.plot(t,v, label = lbl, color = clr)
        plt.grid()
        plt.legend()
        plt.xlabel("time")
        plt.ylabel("amplitude")
        plt.title(tit)
        
    t = np.linspace(0,ts,1000)
    v1 = A*np.sin(2*np.pi*f*t)
    v2 = A*np.sin((2*np.pi*f*t)+(120*np.pi/180))
    v3 = A*np.sin((2*np.pi*f*t)+(240*np.pi/180))
    
    plt.figure(3, figsize = (15,10), dpi = 200)
    
    plt.subplot(2,3,1)
    sin_plot(t, v1, "v1", "Phase A","red")
    
    plt.subplot(2,3,2)
    sin_plot(t, v2, "v2", "Phase B","yellow")
    
    plt.subplot(2,3,3)
    sin_plot(t, v3, "v3", "Phase C","green")
    
    plt.subplot(2,1,2)
    sin_plot(t, v1, "v1", None ,"red")
    sin_plot(t, v2, "v2", None,"yellow")
    sin_plot(t, v3, "v3", "Phase A, B and C","green")

    plt.suptitle(lbl)
    
PrintSine(5,5,0,2,"Three Phase Voltages")
```

```python
def PrintCicle(a,b,r,lbl,clr):
    x = np.linspace(a-r,a+r,100)
  
    y_p = (((r**2)-((x-a)**2))**0.5)+b
    y_n = -(((r**2)-((x-a)**2))**0.5)+b
    
    plt.plot(a,b, marker = "o")
    plt.plot(x,y_p, color = clr, label = lbl)
    plt.plot(x,y_n, color = clr)
    

plt.figure(4, figsize = (10,5), dpi = 100)
PrintCicle(2,15,4,"circle 1","blue")
PrintCicle(15,15,4,"circle 2", "red")
plt.legend()
plt.grid()
```
