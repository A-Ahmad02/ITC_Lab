## Question 1

```python
class Animal(object):
    def __init__(self,name,age):
        
        self.name = name 
        self.age = age
        
    def get_age(self):
        return self.age
    
    def get_name(self):
        return self.name
        
    def set_age(self,newage):
        self.age = newage
    
    def set_name(self,newname):
        self.name = newname
        
    def __str__(self):
        return "animal:"+str(self.name)+":"+str(self.age)
    
class Cat(Animal):
    
    def __init__(self,f,a,b):
        Animal.__init__(self,a,b)
        self.f = f
    
    def speak(self):
        print("meow")
        
    def __str__(self):
        return "cat:"+str(self.name)+":"+str(self.age)
    
    
M = Cat("Yes","Milo",5)
F = Cat("No","fluffy",5)

print(M)
print(F)
M.speak()
print(M.get_age())
print(M.f)
```

## Question 2

```python
class Person(Animal):
    def __init__(self,name,age):
        super().__init__(name,age)
        self.friends = []
        
    def get_friends(self): 
        return self.friends
        
    def add_friend(self,new):
        if new not in self.friends:
            self.friends.append(new)
        
    def speak(self):
        print("hello")
        
    def __str__(self):
        return "person:"+str(self.name)+":"+str(self.age)
    
p1 = Person("jack",30)
p2 = Person("jill",25)
print("p1 name = ", p1.get_name())
print("p2 name = ", p2.get_age())
print(p1)
p1.speak()
print("p1 friends = ",p1.get_friends())
p1.add_friend("john")
print("p1 friends = ",p1.get_friends())
```

## Question 4

```python
class Bird(object):
    def __init__(self,name):
        self.name = name
        self.eggs = 0
        
    def fly(self):
        return "I can fly"
    
    def LayEgg(self):
        self.eggs += 1
    
    def CountEggs(self):
        return self.eggs
    
    def __str__(self):
        return str(self.name)+" has "+str(self.eggs)+" egg(s)"
    
class Penguin(Bird):
    def swim(self):
        return "I can swim."
        
    def fly(self):
        return "No flying for me."
    

print("------Parrot Testing------\n")
bird1 = Bird("Parrot")
print(isinstance(bird1,Bird))
print(bird1.fly())
print("No. of eggs = ", bird1.CountEggs())
print(bird1)
bird1.LayEgg()
print("No. of eggs = ", bird1.CountEggs())
print(bird1)
bird1.LayEgg()
print("No. of eggs = ", bird1.CountEggs())
print(bird1)


print("\n------Penguin Testing------\n")
bird2 = Penguin("Emperor Penguin")
print(isinstance(bird2,Penguin))
print(bird2.fly())
print(bird2.swim())
print("No. of eggs = ", bird2.CountEggs())
print(bird2)
bird2.LayEgg()
print("No. of eggs = ", bird2.CountEggs())
print(bird2)
bird2.LayEgg()
print("No. of eggs = ", bird2.CountEggs())
print(bird2)

```

```python
""" Method 1 - Dictionary - Assuming 2 or more inputs  """

class Gate(object):
    def __init__(self):
        self.inputs = {}
        
    def NumOfInputs(self):
        if len(self.inputs) > 2:
            return len(self.inputs)
        return 2
    
    def setInput(self,num,bool_):
        self.inputs[num] = bool_
        
    def __str__(self):
        list_ = sorted(self.inputs.values())
        for i in range(len(list_)):
            list_[i] = str(list_[i])
        a = ",".join(list_)
        if isinstance(self,AndGate):
            return "And("+a+")"
        return "Or("+a+")"
    
class AndGate(Gate):
    def getOutput(self):
        for i in self.inputs.values():
            if not i:
                return False
        return True 
    
class OrGate(Gate):
    def getOutput(self):
        for i in self.inputs.values():
            if i:
                return True
        return False

print("------And Gate------")
and1 = AndGate()
print(isinstance(and1,AndGate))
print(and1.NumOfInputs())
and1.setInput(1, True)
and1.setInput(2, False)
print(and1)
print("Output = ",and1.getOutput())
and1.setInput(2, True)
print(and1)
print("Output = ",and1.getOutput())

print("------Or Gate------")
or1 = OrGate()
print(isinstance(or1,OrGate))
print(or1.NumOfInputs())
or1.setInput(1, False)
or1.setInput(2, False)
print(or1)
print("Output = ",or1.getOutput())
or1.setInput(2, True)
print(or1)
print("Output = ",or1.getOutput())
```

```python
""" Method 3 - Lists - Assuming 2 or more inputs  """

class Gate(object):
    def __init__(self):
        self.inputs = []
        
    def NumOfInputs(self):
        if len(self.inputs) > 2:
            return len(self.inputs)
        return 2
    
    def setInput(self,num,bool_):
        if num > len(self.inputs):
            self.inputs.append(bool_)
        else:
            self.inputs[num-1] = bool_
        
    def __str__(self):
        list_ = []
        for i in self.inputs:
            list_.append(str(i))
        a = ",".join(list_)
        if isinstance(self,AndGate):
            return "And("+a+")"
        return "Or("+a+")"
    
class AndGate(Gate):
    def getOutput(self):
        for i in self.inputs:
            if not i:
                return False
        return True 
    
class OrGate(Gate):
    def getOutput(self):
        for i in self.inputs:
            if i:
                return True
        return False

print("------And Gate------")
and1 = AndGate()
print(isinstance(and1,AndGate))
print(and1.NumOfInputs())
and1.setInput(1, True)
and1.setInput(2, False)
print(and1)
print("Output = ",and1.getOutput())
and1.setInput(2, True)
print(and1)
print("Output = ",and1.getOutput())

print("------Or Gate------")
or1 = OrGate()
print(isinstance(or1,OrGate))
print(or1.NumOfInputs())
or1.setInput(1, False)
or1.setInput(2, False)
print(or1)
print("Output = ",or1.getOutput())
or1.setInput(2, True)
print(or1)
print("Output = ",or1.getOutput())
```
