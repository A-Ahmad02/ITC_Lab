```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(1,figsize=(15,10)) # Create a figure with tag 1 and width = 15 inches Height = 10 inches
plt.plot([1,2,3],[1,2,3],label='line 1',marker='o',color='red',linestyle='--') #plot in figure 1
plt.xlabel('Integers 1') # Label the x-axis
plt.ylabel('Integers 2') # Label the y-axis
plt.title('A plot')      # Title of the plot
plt.grid()               # Display the grid 
plt.legend()             # Display the legends
```

```python
#Plotting multiple datasets in a single figure

plt.figure(2) # Create a figure with tag 2, Default width and height taken.
plt.plot([1,2,3],[1,2,3],label='line 1',marker='o',color='red',linestyle='--')  # Plot in figure 2
plt.plot([1,2,3],[3,2,1],label='line 2',marker='X',color='green',linestyle='-') # Plot in figure 2
plt.xlabel('Integers 1')
plt.ylabel('Integers 2')
plt.title('A plot')
plt.grid()
plt.legend()
```

```python
#Sub-plotting

plt.figure(3)

plt.subplot(1,2,1)
plt.plot([1,2,3],[1,2,3],label='line 1',marker='o',color='red',linestyle='--')
plt.xlabel('Integers 1')
plt.ylabel('Integers 2')
plt.title('A plot')
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot([1,2,3],[3,2,1],label='line 2',marker='X',color='green',linestyle='-')
plt.xlabel('Integers 1')
plt.ylabel('Integers 2')
plt.title('A plot')
plt.grid()
plt.legend()

plt.suptitle('Subplotting')
```

```python
#Plotting a sine wave

PhaseShift = 90 * (np.pi/180)
t = np.linspace(0,1,41000)
f = 5
v1 = 5 * np.sin(2*np.pi*f*t)
v2 = 1 * np.sin( (2*np.pi*f*t) + PhaseShift)
plt.figure(3)
plt.plot(t,v1,color='red')
plt.plot(t,v2,color='blue')
plt.grid()
```
